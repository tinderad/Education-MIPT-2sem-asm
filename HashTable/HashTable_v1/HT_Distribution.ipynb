```python
import matplotlib
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os import system
```

```python
def drawBars(names):
    fig, ax = plt.subplots()
    for name in names:
        table = pd.read_csv("csv/" + name + ".csv")
        x = table["hash"]
        y = table["quantity"]
        ax.plot(x, y, label= name + ":\n   D = " + str(round(np.var(y),3))+ '\n   avr: '+ str(round(np.average(y),1)))
    plt.grid()
    plt.legend()
    plt.xlabel("hash value")
    plt.ylabel("collisions")
    plt.title(name)
    plt.savefig("plots/" + name + ".pdf")
    plt.show()
```

```python
filenames = [("hash_1",), ("hash_len",), ("hash_sum",), ("hash_sum_len",), ("hash_ascii_0",), ("ded_ror_hash","ded_rol_hash"), ("MeowPurr_hash",), ("crctest",)]

for filename in filenames: drawBars(filename)
```
