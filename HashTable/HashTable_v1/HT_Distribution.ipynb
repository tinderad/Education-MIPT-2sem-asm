```python
import matplotlib
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
plt.rcParams['figure.dpi'] = 300
```

```python
def drawBars(filepath):
    table = pd.read_csv(filepath)
    x = table["hash"]
    y = table["quantity"]
    
    name = filepath[4:-4]
    disp = round(np.var(y),3)
    avr  = round(np.average(y),1)
    
    if len(y) > 5000:
        print("'" + name + "' is too big for plotting. Don't worry, here are:")
        print("  the average:    "+str(avr))
        print("  the dispersion: "+str(disp)+"\n")
        return
    
    fig, ax = plt.subplots()
    ax.bar(x, y, label= name + ":\n   D = " + str(disp)+ '\n   avr: '+ str(avr))
    plt.grid()
    plt.legend()
    plt.xlabel("hash value")
    plt.ylabel("collisions")
    plt.title(name)
    plt.savefig("plots/" + name + ".pdf")
    plt.show()
```

```python
import os

directory = "csv/"

for filename in os.listdir(directory):
    drawBars(os.path.join(directory, filename))
    
```
